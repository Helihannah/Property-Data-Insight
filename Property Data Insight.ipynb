```python
import statsmodels.api as sm

X1 = pd.read_csv("X1.csv")
Y1 = pd.read_csv("Y1.csv")
lm_bad = sm.OLS(Y1, X1)
fit_bad = lm_bad.fit()
```

```python
X2 = pd.get_dummies(X1['OverallQual'], drop_first = True, prefix = 'OverallQual')
display(X2)
```

```python
X3 = pd.get_dummies(X1['OverallCond'], drop_first = True, prefix = 'OverallCond')
display(X3)
```

```python
X4 = pd.get_dummies(X1['KitchenQual'], drop_first = True, prefix = 'KitchenQual')
display(X4)
```

```python
X_11 = pd.merge(X1, X2, left_index=True, right_index=True)
X_11 = pd.merge(X_11, X3, left_index=True, right_index=True)
X_11 = pd.merge(X_11, X4, left_index=True, right_index=True)
```

```python
X_11 = X_11.drop(['OverallCond'], axis=1)
X_11 = X_11.drop(['OverallQual'], axis=1)
X_11 = X_11.drop(['KitchenQual'], axis=1)
```

```python
display(X_11)
```

```python
lm_bad = sm.OLS(Y1, X_11)
fit_bad = lm_bad.fit()
```

```python
fit_bad.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

pd.Series([variance_inflation_factor(X_11.values, i) for i in range(X_11.shape[1])],index=X_11.columns)
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan 
```

```python
result_bp_test = het_breuschpagan(fit_bad.resid, lm_bad.exog)
bp_lm_statistic = result_bp_test[0]
bp_lm_pval = result_bp_test[1]
bp_F_statistic= result_bp_test[2] 
bp_F_pval = result_bp_test[3]
bp_test_output=pd.Series(result_bp_test[0:4],index=['bp_lm_statistic','bp_lm_pval','bp_F_statistic','bp_F_pval'])   
print(bp_test_output)
```
